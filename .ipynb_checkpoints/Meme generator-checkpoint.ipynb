```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
str = input("Enter the name of your text file: ");  

img = cv2.imread(str);
```

```python
# fixing the colors
RGB =  cv2.COLOR_BGR2RGB
plt.figure(figsize = (10,15))
img_rgb = cv2.cvtColor(img, RGB)
```

```python
plt.figure(figsize = (10,15))
print("Here is your image:")
plt.imshow(img_rgb);
```

```python
text_top = input("Enter the text for the TOP of the picture: ").upper();  
```

```python
from PIL import ImageFont, ImageDraw, Image
```

```python
img.shape
```

```python
pil_im = Image.fromarray(img_rgb)
image = img_rgb.copy()

#  creating the draw
draw = ImageDraw.Draw(pil_im)
fontsize = 72
bordercolor = (0,0,0)
fillcolor = (255,255,255)
x_upper = 10
y_upper = 10

x_lower = 2000
y_lower = 200

W, H = img.shape[0], img.shape[1]
msg = "hello"

im = Image.new("RGBA",(W,H),"yellow")
draw = ImageDraw.Draw(im)
w, h = draw.textsize(msg)
draw.text(((W-w)/2,(H-h)/2), msg, fill="black")

im.save("hello.png", "PNG")

# loading the font, with the size
font = ImageFont.truetype("fonts/impact.ttf", fontsize)

draw.text((x_upper-5, y_upper), text_top, font=font, fill=bordercolor)
draw.text((x_upper+5, y_upper), text_top, font=font, fill=bordercolor)
draw.text((x_upper, y_upper-5), text_top, font=font, fill=bordercolor)
draw.text((x_upper, y_upper+5), text_top, font=font, fill=bordercolor)

draw.text((x_lower-5, y_lower), text_bottom, font=font, fill=bordercolor)
draw.text((x_lower+5, y_lower), text_bottom, font=font, fill=bordercolor)
draw.text((x_lower, y_lower-5), text_bottom, font=font, fill=bordercolor)
draw.text((x_lower, y_lower+5), text_bottom, font=font, fill=bordercolor)

draw.text((x_upper, y_upper), text_top, font=font, fill=fillcolor)
draw.text((x_lower, y_lower), text_bottom, font=font, fill=fillcolor)

plt.figure(figsize = (10,15))
plt.imshow(pil_im)
```
